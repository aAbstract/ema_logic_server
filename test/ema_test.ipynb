```python
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
import numpy as np
import math
```

```python
x = np.array([1.0, 2.0, 1.0, 1.0, 1.5])
y = fft(x)
```

```python
print(f'fft output: {y}')
print(f'sum fft values: {np.sum(x)}')
```

```python
for i in range(len(y)):
    print(f'wight for #{i} term: {np.abs(y[i])}')
```

```python
# scipy labeled fourier test
N = 100
T = 1.0
x_fft_test = np.linspace(0.0, T, N)
y_fft_test = [math.sin(2 * math.pi * i) for i in x_fft_test]
plt.plot(x_fft_test, y_fft_test)
```

```python
fft_data = fft(y_fft_test)
print(f'fft data: {len(fft_data)}')
for i in range(len(fft_data[1:50])):
    print(f'wight for #{i + 1} term: {np.abs(fft_data[i + 1])}')
# plt.figure(figsize=(12, 6))
fig, ax = plt.subplots()
fig.set_size_inches(12, 6)
plt.grid()
ax.set_xticks(np.arange(1, 50))
fft_data_to_plot = [np.abs(x) for x in fft_data[1:50]]
plt.plot(range(1, 50), fft_data_to_plot)
```

```python
dom_mag = (2 / N) * np.abs(fft_data[1])
phase_shift = np.angle([fft_data[1]])
print(f'dom_mag: {dom_mag}')
print(f'phase_shift: {phase_shift[0]}')
x_fft_gen_test = np.linspace(0.0, T, N)
thetas = [2 * math.pi / T * theta for theta in x_fft_gen_test]
y_fft_gen_test = [dom_mag * math.cos(point + phase_shift) for point in thetas]
plt.figure(figsize=(12, 6))
# custom inverse
plt.subplot(1, 2, 1)
plt.plot(x_fft_gen_test, y_fft_gen_test)
# auto inverse
plt.subplot(1, 2, 2)
plt.plot(x_fft_gen_test, ifft(fft_data))
print(f'ifft {ifft(fft_data)}')
```
